```python
# prompt: dbscan

from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate sample data
X, y = make_blobs(n_samples=4000, centers=[[4,3],[2,-1],[-1,4]], cluster_std=1.0)


# Plot the clustered data
plt.scatter(X[:, 0], X[:, 1], )
plt.title('DBSCAN Clustering')
plt.show()
```

```python
# Define the DBSCAN model
model = DBSCAN(eps=0.3, min_samples=4)

# Fit the model to the data
y_pred = model.fit_predict(X)
```

```python
# prompt: visual

import matplotlib.pyplot as plt
# Plot the clustered data with different colors for each cluster
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.show()
```

```python
# prompt: kmeans

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Define the KMeans model with 3 clusters
kmeans = KMeans(n_clusters=3)

# Fit the model to the data
y_pred_kmeans = kmeans.fit_predict(X)

# Plot the clustered data with different colors for each cluster
plt.scatter(X[:, 0], X[:, 1], c=y_pred_kmeans, cmap='viridis')
plt.title('KMeans Clustering')
plt.show()
```
