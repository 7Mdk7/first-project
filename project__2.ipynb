# **creating train and test dataset **



```python
import numpy as np

msk = np.random.rand(len(df))<0.8
train = df[msk]
test = df[~msk]
```

**modling**

```python
from sklearn import linear_model
reg=linear_model.LinearRegression()
train_x =np.asanyarray(train[['ENGINESIZE']])
train_y =np.asanyarray(train[['CO2EMISSIONS']])
reg.fit(train_x ,train_y)
```

```python
test_x=np.asanyarray(test[['ENGINESIZE']])
test_y =np.asanyarray(test[['CO2EMISSIONS']])
pridct_y=reg.predict(test_x)
```

**evaluation**

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error, mean_squared_error
print("Mean absolute error :%.2f "% np.mean(np.absolute(pridct_y-test_y)))
print("Mean absolute error :%.2f  "% mean_absolute_error(test_y, pridct_y))
print("residual sum of squares (MES): %.2f"% mean_squared_error(test_y, pridct_y))
print("residual sum of squares (MES): %.2f"%np.mean((pridct_y-test_y)**2))
print("R2-scorw : %.2f "% r2_score(test_y, pridct_y ))
```
