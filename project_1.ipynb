#import data

```python
import pandas as pd
import numpy as np

df1 = pd.read_csv(r'/content/Global Missing Migrants Dataset.csv')
df1.head(10)
```

#Understanding data

```python
df1.describe()
```

```python
df1.isna().sum()
```

#missing data

```python
df1['Region of Origin'].value_counts()
```

```python
df1['Region of Origin'].replace(np.nan,'Latin America / Caribbean (P)',inplace=True)
df1.isna().sum()
```

```python
df1['Country of Origin'].value_counts
```

```python
df1['Country of Origin'].replace('Unknown',np.nan,inplace=True)
```

```python
df1.isna().sum()
```

```python
df = pd.DataFrame({'Country of Origin': ['Mexico', np.nan, 'USA', 'Canada', np.nan]})

# Define the values to replace NaN values
values = ['Mexico','Sudan']

# Replace NaN values with random values
df1['Country of Origin'] = df1['Country of Origin'].fillna(pd.Series(np.random.choice(values, size=len(df1.index))))
df1['Country of Origin'].value_counts()
```

```python
num = df1['Number of Dead'].mean().round(2)
df1['Number of Dead'].replace(np.nan,num,inplace=True)
```

```python
df1['Migration route'].value_counts()
```

```python
df1['Migration route'].replace(np.nan,'US-Mexico border crossing',inplace=True)
```

```python
df1.dropna(subset=['UNSD Geographical Grouping'],axis=0,inplace=True)
```

```python
df1.dropna(subset=['Information Source'],axis=0,inplace=True)
```

```python
df1.dropna(subset=['Coordinates'],axis=0,inplace=True)
```

```python
df1.to_csv('New data')
df2= pd.read_csv("/content/New data")
df2.head()

```

```python
df2.drop('Unnamed: 0',axis=1)
df2
```
